# DISEÑO DE SISTEMAS DE APRENDIZAJE AUTOMÁTICO

### Un proceso iterativo para aplicaciones listas para la producción

Los sitemas de aprendizaje automático son complejos y únicos. Complejos porque constan de muchas componentes diferentes e implican a muchas partes interesadas distintas. únicos porque dependen de datos, que varían enormemente de un caso de uso a otro. En este libro, aprenderá un enfoque holístico para diseñar sistemas de ML que sean fiables, escalables, mantenibles y adaptables a entornos y requisitos empresariales cambiantes.

El autor, Chip Huyen, cofundador de Claypot AI, considera cada decisión de diseño - como la forma de procesar y crear datos de entrenamiento, qué características utilizar, con qué frecuencia colver a entrenar los modelos y qué supervisor- en el contexto de cómo puede ayudar a su sitema en su conjunto alcance sus objetivos. El marco iterativo de este libro utiliza estudios de casos reales respaldados por amplias referencias.

Este libro le ayudará a afrontar situaciones como:
* Ingeniería de datos y elección de las métricas adecuadas para resolver un problema empresarial.
* Automatizar el proceso de desarrollo, evaluación, despliegue y actualiación continuos de los modelos.
* Desarrollar un sistema de vigilancia para detectar y abordar los problemas que sus modelos pueden encontrar en la producción.
* Arquitectura de una plataforma en ML que sirva para distintos casos de uso.
* Desarrollo de sistemas responsables de LD.

### PREFACIO

Desde el primer curso de aprendizaje automático que impartí en Stanford en 2017, muchas personas me han pedido consejo sobre cómo implementar modelos de ML en sus organizacones. Estas preguntas pueden ser genéricas, como "¿Qué modelo debo usar?. ¿Con qué frecuencia debo volver a entrenar mi modelo?. ¿Cómo puedo detectar cambios en la distribución de datos?. ¿Cómo me aseguro de que las características utilizadas durante el entrenamiento son coherentes con las características utilizadas durante la inferencia?.

Estas preguntas también pueden ser específicas, como "Estoy convencido de que pasar de la predicción por lotes a la predicción en línea dará a nuestor modelo un auemtno de rendimiento, pero ¿cómo convenzo a mi jefe para que me deje hacerlo? o "Soy el científico de datos más veterano de mi empresa y hace poco me han encargado crear nuestra primera plataforma de aprendizaje automático; ¿por dóndo empiezo?.

Mi respuesta corat a todas estas preguntas es siempre: "Depende". Mis respuestas largas a menudo implican horas de discusión para entender de dónde viene el que pregunta, lo que realmente está tratando de lograr, y los pros y loscontras de los diferentes enfoques para su caso de uso específico.

Los sistemas de ML son complejos y únicos. Son complejos porque constan de muchos componentes diferentes (algoritmos de ML, datos, lógica empresarial, métricas de evaluación, infraestructura subyacente, etc) e implican muchas partes interesadas difernetes (científicos de datos, ingenieros de ML, líderes empresariales, usuarios, incluso la sociedad en genral). Los sistemas de ML son únicos porque dependen de datos, que varían enormemente de un caso de uso a otro.

Por ejemplo, dos empresas pueden estar en el mismo ámbito (comercio electrónico) y tener el mismo problema que quiere resolver con el ML (sistema de recomendación), pero sus sistemas de ML resultantes pueden tener una arquitectura de modelos diferentes, utilizar conjuntos de características diferentes, evaluarse con métricas diferentes y aportar rendimientos e la inversión diferentes.

Muchas entradas de blog y tutoriales sobre la producción de ML se centran en responder a una pregunta específica. Si bien este enfoque ayuda a transmitir la idea, puede dar la empresión de que es posible considerar cada una de estas cuestiones de forma aislada. En realidad, los cambios en un componente probablemente afectarán a otros componentes. Por lo tanto, es necesario considerar el sisetam en su conjunto a la hora de tomar cualquier decisión de diseño.

Este libro adopta un enfoque holístico de los sistemas de ML. Tiene en cuenta los distintos compoenentes del sistema y los objetivos de las diferentes partes implicadas. El contenido de este libro se ilustra mediante estudios de casos reales, en muchso de los caules he trabajado personalmente, respaldados poramplias referencias y revisados por profesionales del ML tanto del mundo académico como la de la indsutria.

Las secciones que requieren un conocimiento profundo de una determinado tema por ejemplo, procesamiento por lotes frente a procesamiento por flujo, infraestructura de almacenameitno y computación, e IA responsable- son revisadas por expertos cuyo trabajo se centra en ese tema. En otras palabras, este libro es un intento de dar respuestas matizadas a las preguntas antes mnecionadas y a otras más.


Cuando escribí por primera vez los apuntes de clase que sentaron las bases de este libro, pensé que los escribía para mis alumnos, para prepararlos para las exigencias de sus futuros trabajos como cientifpicos de datos e ingenieros de ML. Sin embargo, pronto me di cuenta de que yo también había parendido muchísimo durante el proceso. Los borrados res iniciales que compartí con los primeros lectores suscitaron muchas conversaciones que pusieron a prueba mis suposiciones, me obligaron a considerar diferentes perspectivas y me presentaron nuevos problemas y enfoques.

Espero que este proceso de aprendizaje continúe para mí ahora que el libro está en tus manos, ya que tiene experiencias y perspetivas que son únicas para ti. Por favor, no dudes en compartir conmigo cualquier comentario que tengas para este libro a través del servidor MLOps Discord que dirigo (donde también puedes encontrar a otros lectores de este libro), Twitter, LinkedIn u otros canales que puede encontrar en mi página web.

### A quién va dirigido este libro

Este libro está dirigido a cualquiera que desee aprovechar el ML para resolver problemas del mundo real. ML en este libro se refiere tanto al aprendizaje profundo como a los algoritmos clásicos, con una inclinación hacia los sitemas de ML a escala, como los que se ven en empresas mediana y grandes y en startups de rápido crecimiento. Los sistemas a menor escala tienden a ser menos complejos y podrían beneficiarse menos del enfoque integral que se expone en este libro.

Dado que mi formación es de ingeniería, el lenguaje de este libro está dirigido a ingenieros, incluidos ingenieros de ML, científicos de datos, ingenieros de datos, ingenieros de plataformas de ML y directores de ingeniería. Puede que te sientas identificado con alguno de los siguientes escenarios:

- Se le ha planteado un problema empresarial y se le ha proporcionado muchos datos en bruto. Quieres diseñar estos datos y elegir las métrcias adecuadas para resolver el problema.
- Sus modelos iniciales funcionan bien en experimentos fuera de línea y quiere implementarlos.
- Tiene poca información sobre el rendimiento de sus modelos una vez desplegados, y quiere encontrar una forma de detectar, depurar y solucionar rápidamente cualquier problema que puedan tener sus modelos en producción.
- El proceso de desarrollo, evaluación , despliegue y actualización de modelos para su equipo ha sido en su mayor parte manual, lento y propenso a erores. Quiere automatizar y mejorar este proceso.
- Cada caso de uso de ML en su organización se ha desplegado utilizando su propio flujo de trabajo y usted desea sentar las baes (p.ej,. almace´n de modelos, almacén de funciones, herramientas de supervisión) que puedan compartirse y reutilizarse en todos los casos de uso.

También puede beneficiarse del libro si pertenece a uno de los siguientes grupos:

### Lo que este libro no es

Este libro no es una introducción al ML. Hay muchos libros, cursos y recursos disponibles para las teorías de ML, y por lo lanto, este libro se elja de estos conceptos para centrarse en los aspectos prácticos de ML. Para ser específicos, el libro asume que los lectores tiene un compresnión básica de los siguientes temas:

- Modelos de ML como clustering, regresión logística, árboles de decisión, filtrado colaborativo y diversas arquitecturas de redes neuronales como feed-forward, recur-rent, convolucionales y transformadoras.
- Técnicas de ML como la supervisión frente a la no supervisión, el descenso de gradiente, la función objetivo/pérdido, la regularización, la generalización y el ajuste de hiperparámetros.
- Métricas como exactitud, F1, precisión, recuerdo, ROC, error cuadrático medio y log-versimilitud.
- Conceptos estadísticos como varianza, probabilidad y distribución normal o de larga cola.
- Tareas habituales de ML, como el modelo del lenguaje, la detección de anoamalías, la clasificación de objetos y la traducción automática.

No es ncesario que conozca estos temas al dedillo - para conceptso cuyas definiciones exactas pueden requerir cierto esfuerzo para recorda, por ejemplo, la puntuación F1, incluimos breves notas a modo de referencia, pero debe tener una idea aproximada de lo que significan.

Aunque este libro menciona herramientas actuales para ilustrar ciertos conceptos y soluciones, no es un libro tutorial. Las tecnologpias evoluacionan con el tiempo. Las herramientas entran y salen de moda rápidamente, pero los enfoques fundamentales para la resolución deproblemas deberían duran un poco más. Este libro proporciona un marco para que evalúe la herramienta que mejor se adapta a sus casos de suo. Cuando hay una herramienta que quieres utilizar, suele ser fácil encontrar tutoriales en línea. Como resultado, este libro contiene pocos fragmentos de código y se centra en proporcionar una gran cantidad de discusión e torno a compensaciones, pero y contras, y ejemplos concretos.


### Navegar por este libro

Los capítulos de este libro estpan organizados para reflejar los problemas que los científicos de daots pueden encontrar a medida que avanzan en el ciclo de vida de un proyecto de ML. Los dos primers capítulos sientan las bases para que un proyecto de ML tenga éxito, empezando por la pregunta más básica: `¿necesita su proyecto ML?` También se aborda la elección de los objetivos del proyecto y cómo plantear el problema de forma que las solucions sean más sencillas. Si ya está familiarizado con estas consideraciones e impaciente por llegar a las coluciones técnicas, no dude en saltarse los dos primeros capítulos.

Los capítulos 4 a 6 cubren la fase previa al despligue de un proyecto de ML: desde la creación de los datos de entrenamiento y las caracter´siticas de ingeniería hasta el desarrollo y la evaluación de sus modelos en un entorno de desarrollo. Esta es la fase en la que son especialemnte necesario los conocimientos tanto de ML como del dominio del problema.

Los capítulo 7 a 9 cubren las fases de despliegue y post-despliegue de un proyecto de ML. A través de una historio con la que muchso lectores se sentiran identificados, aprenderemos que el despliegue de un modelo no es el final del proceso de despliegue. El modelo desplegado deberá supervisarse y actualizarse continuamente en función de los cambios del entorno y los requisitos empresariales.

Los capitulos 3 y 10 se centran en la infraestructura necesaria para que las partes interesadas de distintos ámbitos puedan trabajar juntas y ofrecer sistemas e ML de éxito. El capítulo 3 se centra en los sistemas de datos, mientras que el capítulo 10 se centra en la infraestructura informática y las plataformas de ML. He debatido durante mucho tiempo hasta qué punto profundizar en los sitemas de datos y dónde introducirlos en el libro. Sistemas de datos, icnluyendo bases de datos, formatos de datos, los movimientos de datos y los motores de procesamiento de datos suelen tratarse poco en los crusos de ML, por lo que muchos científicos de datos podrían considerarlo de bajo nivel o irrelevantes. Después de consultar con muchos de mis colegas, decidí que, dado que los sistemas de ML depende de los datos, cubrir los fundamentos de los sistemas de datos desde el principio nos ayudará a ponernos de acuerdo para discutir asuntos de datos en el resto del libro.

Aunque en este libro cubrimos muchos aspectos técnicos de un sistema de ML, los sitemas de ML están construidos por personas, para personas, y pueden tener un impacto enorme en la vida de muchas personas. Sería negligente escribir un libro sobre la producción de ML sin un capítulo sobre su lado humano, que es el tema cantral del capítulo 11, el último.

Ten en cuenta que "científicos de datos" es una función que ha evolucionado mucho en los últimos años, y ha habido muchos debates para determinar lo que debería implicar este función. En este libro, utilizamos "cientifico de datos" como término general para incluir a cualquiera que trabaje desarrollando e implementando modelos de ML, incluidas las personas cuyos títulos laborales podrían ser ingenieros de ML, ingenieron de daos, analistas de datos, etc.
